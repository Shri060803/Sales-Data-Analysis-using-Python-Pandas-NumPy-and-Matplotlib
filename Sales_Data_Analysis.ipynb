```python
sales_by_year = df.groupby('Year')['Sales'].sum()
```

```python
sales_by_year
```

```python
sales_by_year.plot(kind='bar')
```

```python
for year, sales in sales_by_year.items():
    if sales >= 10000000:  # 1 Crore
        print(f"{year}: {sales/10000000:.2f} Crore")
    else:
        print(f"{year}: {sales/100000:.2f} Lakh")
```

```python
profit_by_year = df.groupby('Year')['Profit'].sum()
profit_by_year
```

```python
profit_by_year.plot(kind='bar')
```

```python
city_sales = df.groupby("City")["Sales"].sum().sort_values(ascending=False).head(10)

city_sales.plot(kind="bar", figsize=(8,4), title="Top 10 Cities by Sales")
```

```python
category_sales = df.groupby("Category")["Sales"].sum()

category_sales.plot(kind="pie", autopct="%1.1f%%", figsize=(6,6), title="Sales by Category")
```

```python
profit_category = df.groupby("Category")["Profit"].sum()

profit_category.plot(kind="bar", figsize=(5,3), title="Profit by Category")
```

```python
top_products = df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10)

top_products.plot(kind="bar", figsize=(8,3), title="Top 10 Products by Sales")
```
